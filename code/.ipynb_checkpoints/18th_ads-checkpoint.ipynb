```python
import xml.etree.ElementTree as ET
tree = ET.parse('../data/ads_tei_edit.xml')
root = tree.getroot()
```

```python
root.tag
```

```python
for child in root:
    print(child.tag, child.attrib)

```

```python
for heading in root.findall('body/div'):
    head = heading.find('p/hi').text
    print(head)
```
